```python
import tensorflow as tf
import pandas as pd
```

```python
A=pd.read_csv("50_Startups.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot(A['Profit'])
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
SC=StandardScaler
LB=LabelEncoder()
```

```python
A['State']=LB.fit_transform(A['State'])
```

```python
A.head()
```

```python
X=A.drop(['Profit'],axis=1)
Y=A.Profit
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.80,random_state=50)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
Sd=StandardScaler()
```

```python
X_train=Sd.fit_transform(X_train)
X_test=Sd.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
SQ=Sequential()
```

```python
## input layer 
SQ.add(Dense(units=40,activation='ReLU'))
```

```python
## hidden layer 1 
SQ.add(Dense(units=50,activation='ReLU'))
SQ.add(Dropout(0.3))
```

```python
## hidden layer 2
SQ.add(Dense(units=45,activation='LeakyReLU'))
SQ.add(Dropout(0.2))
```

```python
## hidden layer 3
SQ.add(Dense(units=40,activation='ReLU'))
SQ.add(Dropout(0.1))
```

```python
SQ.add(Dense(units=1,activation='linear'))
```

```python
SQ.compile(optimizer='adam', loss='Huber', metrics=['mse'])
```

```python
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
)
```

```python
model_1=SQ.fit(X_train,Y_train,validation_split=0.33,batch_size=10,epochs=1000,callbacks=early_stopping)
```

```python
model_1.history.keys()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot(model_1.history['loss'])
plt.plot(model_1.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(model_1.history['mse'])
plt.plot(model_1.history['val_mse'])
plt.ylabel('mse')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```
